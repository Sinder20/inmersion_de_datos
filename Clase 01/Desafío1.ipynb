## Desafío:
1. Promedio del área de todos los inmuebles en los barrios en el dataset. El top 10;

```python
import pandas as pd

inmuebles = pd.read_csv("inmuebles_bogota.csv")
inmuebles.head()
```

```python
columnas = {'Baños':'Banos','Área':'Area'}
inmuebles = inmuebles.rename(columns=columnas)
inmuebles
```

```python
inmuebles_prom_area_barrios = inmuebles.groupby('Barrio').Area.mean().sort_values(ascending=False)
inmuebles_prom_area_barrios.head(10)
```

```python
inmuebles_prom_area_barrios.head(10).plot.bar(color='magenta')
```
